```python
import numpy as np
import matplotlib.pyplot as plt
print('Obtain Gaussian Random Samples')
U =[[1/np.sqrt(2), -1/np.sqrt(2)], [1/np.sqrt(2), 1/np.sqrt(2)]]
U = np.asarray(U)

Lambda_1 = [[4,0],[0,1]]
Labmda_1 = np.asarray(Lambda_1)
Lambda_2 = [[1,0],[0,5]]
Labmda_2 = np.asarray(Lambda_2)

Cov_1 = U.dot(Lambda_1).dot(U.T)
Cov_2 = U.dot(Lambda_2).dot(U.T)

X_1 = np.random.multivariate_normal([1, 0], Cov_1, 1000).T
X_2 = np.random.multivariate_normal([-1, 0], Cov_2, 1000).T
X_3 = np.random.multivariate_normal([0, -10], np.identity(2), 1000).T
X_1_orig = X_1.T; X_2_orig = X_2.T; X_3_orig = X_3.T
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_1[0], X_1[1], c= 'r')
ax.scatter(X_2[0], X_2[1], c = 'g')
ax.scatter(X_3[0], X_3[1], c = 'b')
plt.show()
```

```python
print('randomize the samples')

X_big = X_1_orig.tolist()
X_big.extend(X_2_orig.tolist())
X_big.extend(X_3_orig.tolist())

#X_big_index = [tuple(sample) for sample in X_big]

X_big_index = np.random.choice(a=range(3000), size=3000, replace=False)

X_big_rand = []
for index in X_big_index.T:
    X_big_rand.append(X_big[index])

X_1_orig = np.asarray(X_big_rand[:1000]).T; X_2_orig = np.asarray(X_big_rand[1000:2000]).T; X_3_orig = np.asarray(X_big_rand[2000:]).T;
#X_1_rand = np.asarray(X_rand[:1000]);X_2_rand = np.asarray(X_rand[1000:2000]);X_3_rand = np.asarray(X_rand[2000:]);
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_1_orig[0], X_1_orig[1], c= 'r')
ax.scatter(X_2_orig[0], X_2_orig[1], c = 'g')
ax.scatter(X_3_orig[0], X_3_orig[1], c = 'b')
plt.show()
```

```python
X_1 = X_1_orig; X_2 = X_2_orig; X_3 = X_3_orig

print('Clustering')

#Calculate mean
X_1_mean = np.mean(X_1, axis = 1)
X_2_mean = np.mean(X_2, axis = 1)
X_3_mean = np.mean(X_3, axis = 1)
#print(X_1_mean)
#Find the distances
X_1_new = []; X_2_new = []; X_3_new = []

isFinished = False
counter = 0
while((not isFinished) and counter <= 200):    
    isFinished = True
    numDiff = 0
    X_1_mean = np.mean(X_1, axis = 1)
    X_2_mean = np.mean(X_2, axis = 1)
    X_3_mean = np.mean(X_3, axis = 1)
    for i in range(len(X_1.T)):

        sample = X_1[:,i]
        dist_1, dist_2, dist_3 = (np.linalg.norm(sample - X_1_mean), 
                                  np.linalg.norm(sample - X_2_mean), 
                                  np.linalg.norm(sample - X_3_mean))
        group = np.argmin([dist_1, dist_2, dist_3])
        
        #print(group)
        if(group == 0):
            X_1_new.append(sample.tolist())
        elif(group == 1):
            isFinished = False;
            numDiff +=1
            X_2_new.append(sample.tolist())
        elif(group == 2):
            isFinished = False;
            numDiff +=1
            X_3_new.append(sample.tolist())

    for i in range(len(X_2.T)):

        sample = X_2[:,i]
        dist_1, dist_2, dist_3 = (np.linalg.norm(sample - X_1_mean), 
                                  np.linalg.norm(sample - X_2_mean), 
                                  np.linalg.norm(sample - X_3_mean))
        group = np.argmin([dist_1, dist_2, dist_3])
        #print(group)
        if(group == 0):
            isFinished = False;
            numDiff +=1
            X_1_new.append(sample.tolist())
        elif(group == 1):
            X_2_new.append(sample.tolist())
        elif(group == 2):
            isFinished = False;
            numDiff +=1
            X_3_new.append(sample.tolist())

    for i in range(len(X_3.T)):

        sample = X_3[:,i]
        dist_1, dist_2, dist_3 = (np.linalg.norm(sample - X_1_mean), 
                                  np.linalg.norm(sample - X_2_mean), 
                                  np.linalg.norm(sample - X_3_mean))
        group = np.argmin([dist_1, dist_2, dist_3])
        #print(group)
        if(group == 0):
            isFinished = False;
            numDiff +=1
            X_1_new.append(sample.tolist())
        elif(group == 1):
            isFinished = False;
            numDiff +=1
            X_2_new.append(sample.tolist())
        elif(group == 2):
            X_3_new.append(sample.tolist())
    
    X_1 = np.asarray(X_1_new).T; X_2 = np.asarray(X_2_new).T; X_3 = np.asarray(X_3_new).T
    X_1_new = []; X_2_new = []; X_3_new = [];
    
    print(numDiff)
    counter += 1

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(X_1[0], X_1[1], c= 'r')
ax.scatter(X_2[0], X_2[1], c = 'g')
ax.scatter(X_3[0], X_3[1], c = 'b')
plt.show()
```
